#Question 3: Differentiate between function and method?
'''Answer: A function is block of code that is design to perform a specific task within the program. Method is also a function but it is defined within the class of a program, particularly defines the behavior of an object of a class.'''

#Question 4: Define the following terms: 1. Class 2. Object 3. Attribute 4. Behavior
'''Answer:
1)CLASS: A class is a block of code that contains the attributes and behavior of an object.
2)OBJECT: An object is any real world entity which has its general attributes and behavior.
3)ATTRIBUTE: An attribute is a data of the object.(e.g, for a person class; name, age , gender are the attributes)
4)BEHAVIOR: Behavior of an object defines how the instance of that object operates.(e.g, for a person class; eat, walk, talk etc. are the behaviors)''' 

```python
#Question 5: Write a code in python in which create a class named it Car which have 5 attributes 
#such like (model, color and name etc.) and 3 methods. And create 5 object instance from that class.
class car():
    def __init__(self,name,model,color,no_of_doors,no_of_tyres):
        self.name = name
        self.model = model
        self.color = color
        self.no_of_doors = no_of_doors
        self.no_of_tyres = no_of_tyres
        
    def setcolor(self,newColor):
        self.color = newColor
    def updatemodel(self,newModel):
        self.model = newModel
    def changeCar(self,newName,newModel,newColor,doors,tyres):
        self.name = newName
        self.model = newModel
        self.color = newColor
        self.no_of_doors = doors
        self.no_of_tyres = tyres

car1=car('Koenigsegg','ABC','Red',4,4)
print(car1.name)
print(car1.model)
car1.setcolor('Black')
print(car1.color)
print(car1.no_of_doors)
print(car1.no_of_tyres)
print()
car2=car('Lamborghini','XYZ','Blue',4,4)
print(car2.name)
print(car2.model)
print(car2.color)
print(car2.no_of_doors)
print(car2.no_of_tyres)
print()
car3=car('Ferrari ','JKL','Yellow',4,4)
print(car3.name)
car3.updatemodel('JKL1a')
print(car3.model)
print(car3.color)
print(car3.no_of_doors)
print(car3.no_of_tyres)
print()
car4=car('Mercedes-AMG One ','PQR','Green',4,4)
print(car4.name)
print(car4.model)
car4.setcolor('White')
print(car4.color)
print(car4.no_of_doors)
print(car4.no_of_tyres)
print()
car5=car('Bugatti Divo','LMN','Silver',4,4)
car5.changeCar('BUGATTI LA VOITURE NOIRE','ASD','Black',4,4)
print(car5.name)
print(car5.model)
print(car5.color)
print(car5.no_of_doors)
print(car5.no_of_tyres)
        
```
